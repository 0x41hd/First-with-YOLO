```python
from imageai.Detection import ObjectDetection, VideoObjectDetection
import os
```

```python
model_path = "C:\\Users\\ahdco\\OneDrive\\Desktop\\first-yolo\\First-with-YOLO"
image = "C:\\Users\\ahdco\\OneDrive\\Desktop\\first-yolo\\First-with-YOLO\\image"
video = "C:\\Users\\ahdco\\OneDrive\\Desktop\\first-yolo\\First-with-YOLO\\video"
```

```python
detector = ObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath(os.path.join(model_path, "yolov3.pt"))
detector.loadModel()
```

```python
detection = detector.detectObjectsFromImage(input_image=os.path.join(image, "max.jpg"),output_image_path=os.path.join(image, "max_detect.jpg"))
```

```python
custom_detection = detector.detectObjectsFromImage(input_image= os.path.join(image, "cars.jpg"), output_image_path= os.path.join(image, "car_dete.jpg"), minimum_percentage_probability= 92)
```

```python
video_detector = VideoObjectDetection()
video_detector.setModelTypeAsYOLOv3()
video_detector.setModelPath(os.path.join(model_path, "yolov3.pt"))
video_detector.loadModel()
```

```python
video_detector = video_detector.detectObjectsFromVideo(input_file_path= os.path.join(video, "times.mp4"), output_file_path=  os.path.join(video, "times_detev"), frames_per_second= 30, log_progress= True)
```
